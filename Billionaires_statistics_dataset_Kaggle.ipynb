# Pie Chart of Industry Distribution

```python
import seaborn as sns
import seaborn as sns
import matplotlib.pyplot as plt

from matplotlib.colors import ListedColormap, LinearSegmentedColormap

plt.figure(figsize=(12, 12))

industry_counts = billionaires_data['industry'].value_counts()
percentages = industry_counts / industry_counts.sum() * 100
custom_labels = [f"{labels} ({percentage:.1f}%)" for labels, percentage in zip(industry_counts.index, percentages)]

#create custom color palette

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
          '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf',
          '#aec7e8', '#ffbb78', '#98df8a', '#ff9896', '#c5b0d5',
          '#c49c94', '#f7b6d2', '#c7c7c7']

#add borders to the pie chart
wedgeprops={'edgecolor':'white','linewidth':1}

plt.pie(industry_counts, labels=None, autopct='', startangle=300,colors=colors,wedgeprops=wedgeprops)
plt.title('Pie chart of Industry Distribution', fontsize=14)

# Create a new list containing the column names and percentages
labels_with_percentages = [f"{label}: {percentage:.1f}%" for label, percentage in zip(industry_counts.index, percentages)]

# Display the list outside the pie chart using plt.legend()
plt.legend(labels_with_percentages, loc='upper left', fontsize=10)
plt.tight_layout()
plt.show()
```

# Bar Chart of countries of residence of Billionaires

```python
countries= billionaires_data['country_of_residence'].value_counts()
countries_count=countries.reset_index()
countries_count.columns=['country','counts']
top_10_countries_of_billionaires=countries_count.head(10)
print(top_10_countries_of_billionaires)
```

```python
import matplotlib.pyplot as plt
colors = sns.color_palette("husl", len(top_10_countries_of_billionaires))
plt.figure(figsize=(8, 6))
barplot_two =sns.barplot(x='counts', y='country', data=top_10_countries_of_billionaires,palette=colors)

for p in barplot_two.patches:
    width = p.get_width()
    plt.text(width + 3,
             p.get_y() + p.get_height() / 2,
             f'{int(width)}',
             va='center')
plt.xlabel('Counts')
plt.ylabel('Country of Residence')
plt.title('Bar Chart of Coutries of Residence of Billionaires')
plt.show()
```

# Age Grouping

```python
#age grouping
billionaires_data['age']=pd.to_numeric(billionaires_data['age'],errors='coerce')
#drop rows with missing age values
billionaires_data.dropna(subset=['age'],inplace=True)
#grouping by age
age_groups=billionaires_data.groupby(pd.cut(billionaires_data['age'],bins=range(0,101,10),right=False)).size()
#visualization
plt.figure(figsize=(10,6))
age_groups.plot(kind='bar',color='skyblue')
plt.title('Number of Billionaires by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Billionaires')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
